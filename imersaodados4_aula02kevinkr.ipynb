```python
#IMPORTANDO BIBLIOTECAS 'seaborn' e 'matplotlib'
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

def millions_formatter(x, pos):
    return '%1.1fM' % (x * 1e-6)

#HISTOGRAMA PERSONALIZADO
sns.set()
sns.set_style("white")
plt.figure(figsize = (10,8))
ax = sns.histplot(data = dados_vendas, x = "Valor_anuncio_float", kde = True,edgecolor = 'black', color = "green")
ax.set_title('Histograma Valor Imóvel', fontsize = 20)
plt.xlabel('Valor do Anúncio', fontsize = 15)
plt.ylabel('Qtd de Imóveis', fontsize = 15)
plt.xlim(-50, 10000000)
ax.xaxis.set_major_formatter(FuncFormatter(millions_formatter))

plt.show()
```
