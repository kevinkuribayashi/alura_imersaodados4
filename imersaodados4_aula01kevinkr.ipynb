#### ANÁLISE DESCRITIVA

Com o desvio padrão alto ('std'), é possível identificar que há dispersão grande dos valores de metragem dos imóveis, ou seja, os valores podem se destoar entre eles e longe da média. Comparado a 'Metragem' com as demais variáveis é possível identificar que os valores se aproximam, uma vez que possuem desvio padrão menor.

#### REFERÊNCIAS: DESVIO PADRÃO
Um grande desvio padrão indica que os pontos dos dados estão espalhados longe da média e um pequeno desvio padrão indica que os pontos dos dados estão agrupados perto da média

### 5 - Descobrir quais são os bairros que não tem nome de Rua [Vivi]

```python
dados.isnull().sum() #QUANTIDADE DE REGISTROS NULOS NO DATASET POR VARIÁVEL
#REFERÊNCIAS: https://pandas.pydata.org/docs/reference/api/pandas.isnull.html
```

```python
dados['Rua'].isnull() #TESTANDO .isnull()
#REFERÊNCIAS: https://pandas.pydata.org/docs/reference/api/pandas.isnull.html
```

```python
bairros_semrua = dados[dados['Rua'].isnull()] #SELECIONANDO OS REGISTROS QUE NÃO POSSUEM ENDEREÇO
bairros_semrua.head(20) #TOP20 REGISTROS QUE NÃO POSSUEM ENDEREÇO
#REFERÊNCIAS: https://pandas.pydata.org/docs/reference/api/pandas.isnull.html
```

```python
pd.options.display.max_rows = 10 #SETANDO PARA QUE APRESENTE TODAS O RESULTADO COM TODAS AS LINHAS
bairros_semrua['Bairro'].value_counts() #CONTANDO O NÚMERO DE REGISTROS QUE NÃO POSSUEM ENDEREÇO POR BAIRRO

#REFERÊNCIA: 
# https://stackoverflow.com/questions/46554597/how-to-display-all-rows-in-jupyter-notebook
# https://pandas.pydata.org/docs/reference/api/pandas.Series.value_counts.html
```
